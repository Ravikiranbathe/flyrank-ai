## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My chosen lane is Refresh / Content Opportunity Scoring.

I would frame this as a ranking/scoring problem. The goal is to give content items an opportunity score and rank them so that pages that may need review or refresh appear first.

This would help a content editor decide which pages to review first, instead of manually checking every page.

A wrong ranking could waste the editor's time on a page that was not actually a useful refresh opportunity.

```python
import pandas as pd 

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

For this lane, the target would be a future content decline or refresh opportunity that helps identify which pages should be reviewed first.

The starter dataset has trend_direction and trend_pct, but trend_direction is derived from trend_pct, so they describe the current trend rather than an independent future outcome. So I would use them only as a proxy for exploring refresh opportunities, not as a true future outcome.

For a stronger model later, I would prefer an observed outcome from a later time period.

```python
print(df["trend_direction"].value_counts())
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

I would use **Precision@K** as the main success metric.

This measures how many of the top K pages suggested by the scoring system are actually useful refresh opportunities. This fits the task because an editor would normally review only the highest-priority pages, not every page in the dataset.

A good result would be a Precision@K that is clearly better than a simple baseline or rule-based ranking.


```python
down_rate = (df["trend_direction"] == "down").mean()

print("Down trend rate:", round(down_rate, 3))
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

The unit of analysis is **one content item per row**.

For this lane, each row represents one pseudonymized page/content item that could be scored and prioritized for review. I will look at its recent performance and content-related signals to support the refresh opportunity decision.


```python
lane_df = df[["content_id","content_type","impressions_90d","clicks_90d","sessions_90d","trend_direction"]]

lane_df.head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule may only check one condition, like low clicks or a downward trend. But a page may need a refresh because of different factors such as impressions, clicks, sessions, content type, and search demand.

Since these factors can be different for every content item, one fixed rule may not work well for all pages. ML can use multiple signals together and help rank the pages that may be worth reviewing first.

The final ranking would help the content editor make a decision, rather than automatically deciding which page must be refreshed.

```python
down_pages = df[df["trend_direction"] == "down"]

down_pages[["impressions_90d", "clicks_90d", "sessions_90d"]].describe()
```
